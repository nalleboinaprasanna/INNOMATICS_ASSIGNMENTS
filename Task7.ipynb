```python
#1
#define a class
class Smartlight:
#constructor
    def __init__(self,name):
        self.name=name
        self.status="OFF"
    def turn_on(self):
        self.status="ON"
        print(f"{self.name} is {self.status}")
    def turn_off(self):
        self.status="OFF"
        print(f"{self.name} is {self.status}")
#creating an object for class
light=Smartlight("Bedroom Light")
light.turn_on()
light.turn_off()
```

```python
#2
class Employee:
#constructor
    def __init__(self,name,emp_id,department):
        self.name=name
        self.emp_id=emp_id
        self.department=department
#function for details
    def display(self):
        print("Employee ID Card")
        print("Name:",self.name)
        print("ID:",self.emp_id)
        print("Department:",self.department)
#dynamic input from user
name=input(" Enter Employee Name:")
emp_id=input("Enter Employee ID:")
department=input("Enter Department:")
#obect creation and passsing parameters
emp=Employee(name,emp_id,department)
#function calling
emp.display()
```

```python
#3
class Contact:
#constructor
    def __init__(self,name,phno):
        self.name=name
        self.phno=phno
#function for details
    def display(self):
        print("Contact Saved")
        print("Name:",self.name)
        print("Phone",self.phno)
#dynamic input from user
name=input(" Enter contact Name:")
phno=input("Enter Phone number:")
#obect creation and passsing parameters
c=Contact(name,phno)
#function calling
c.display()
```

```python
#4
#defining a class
class Product:
    def __init__(self,name,price):
        self.name=name
        self.price=price
#funtion for printing
    def print(self):
        print("Product:",self.name)
        print("Price:₹"+str(self.price))
#taking dynamic input from user
name=input("Enter product name:")
price=int(input("Ente price:"))
#object creation
p=Product(name,price)
#calling
p.print()
```

```python
#5
class Movie:
    def __init__(self,name,rating):
        self.name=name
        self.rating=rating
#funtion for printing
    def display(self):
        print("Movie:",self.name)
        print("Rating",self.rating,"/5")
#taking dynamic input from user
name=input("Enter movie name:")
rating=float(input("Enter rating:"))
#object creation
m=Movie(name,rating)
#calling
m.display()
```

```python
#6
class Delivery:
    def __init__(self,name,address):
        self.name=name
        self.address=address
#define a print funtion
    def print_details(self):
        print("Delivery Details")
        print("Customer:",self.name)
        print("Address:",self.address)
#dynamic input from user
name=input("Enter Customer Name:")
address=input("Enter Address:")
#object creation and passing parameters
d=Delivery(name,address)
#calling fucntion
d.print_details()
```
