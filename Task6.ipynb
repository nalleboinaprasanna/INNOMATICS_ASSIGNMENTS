```python
# 1. Getting the parking capacity
capacity=int(input("Enter parking capacity: "))

# 2. Getting the logs
user_input=input("Enter vehicle logs (separated by spaces): ")
logs=user_input.upper().split()
current_parked=0
peak_usage=0

for action in logs:
    if action=="IN":
        current_parked=current_parked+1
    elif action=="OUT":
        current_parked=current_parked-1

    # Tracking the highest number reached (Peak Usage)
    if current_parked>peak_usage:
        peak_usage=current_parked

# 6. Checking for capacity alerts
if current_parked>=capacity:
    status="Full"
else:
    status="Available"
print("Currently Parked Vehicles:", current_parked)
print("Peak Parking Usage Today:", peak_usage)
print("Parking Status:", status)
```

```python
# Creating a function to calculate the delivery time
def delivery(distance, traffic, weather):
    base_time=distance * 2
    if traffic=="HIGH":
        traffic_delay=20
    elif traffic=="MEDIUM":
        traffic_delay=10
    else:
        traffic_delay=0
    if weather=="RAINY":
        weather_delay=15
    elif weather=="SNOWY":
        weather_delay=25
    else:
        weather_delay=0

    # 4. Calculating the final total time
    total_time=base_time+traffic_delay+weather_delay
dist=float(input("Enter Distance (km): "))
traf=input("Enter Traffic Level (Low, Medium, High): ").upper()
weath=input("Enter Weather (Clear, Rainy, Snowy): ").upper()
final_eta=delivery(dist, traf, weath)

# Displaying the final result
print("Estimated Delivery Time:", final_eta, "minutes")
```

```python
# Creating a function
def analyze(total_seats, booked_list):
    booked_count=len(booked_list)
    occupancy=(booked_count / total_seats) * 100

    if occupancy==100:
        status="Housefull"
        suggestion="Opening an additional show is highly recommended!"
    elif occupancy>=80:
        status="Almost Full"
        suggestion="Consider opening a new show soon."
    else:
        status="Seats Available"
        suggestion="No need for additional shows yet."

    return occupancy, status, suggestion

# main program
total=int(input("Enter Total Seats in Theatre: "))
booked_input=input("Enter booked seats: ")
booked_list=booked_input.split()

# Calling the function and unpacking the results
perc, stat, advice=analyze(total, booked_list)

# Printing the results
print("Occupancy:", perc, "%")
print("Show Status:", stat)
print("Suggestion:", advice)
```

```python
# Creating a function
def server_load(cpu_list):
    total_cpu=sum(cpu_list)
    reading_count=len(cpu_list)
    average_load=total_cpu / reading_count
    if average_load<50:
        status="Normal"
    elif average_load<=80:
        status="Warning"
    else:
        status="Critical"
    return average_load, status
user_entry=input("Enter CPU readings separated : ")
cpu_readings=[]
for x in user_entry.split():
    cpu_readings.append(float(x))

# 5. Calling the function and storing the results
avg, server_status=server_load(cpu_readings)

# 6. Displaying the final result
print("Average CPU Load:", avg, "%")
print("Server Status:", server_status)
```

```python
#Creating a function
def monitor_usage(usage_dict):
    overused_list=[]
    alert="No"
    for resource, hours in usage_dict.items():
        if hours>8:
            overused_list.append(resource)
            alert="Yes"

    return overused_list, alert
classroom_data = {}
print("Enter usage hours for the following resources:")
classroom_data["Projector"]=float(input("Projector hours: "))
classroom_data["AC"]=float(input("AC hours: "))
classroom_data["Lights"]=float(input("Lights hours: "))

# Calling the function and unpacking the returned results
overused, energy_alert = monitor_usage(classroom_data)
# Using join to display the list of overused items neatly
print("Overused Resources:", ", ".join(overused) if overused else "None")
print("Energy Alert:", energy_alert)
```

```python
# Creating a function to manage
def registrations(capacity, total_signups):
    # Checking if the number of signups is more than the capacity
    if total_signups>capacity:
        confirmed=capacity
        waitlisted=total_signups - capacity
        status="Closed"
    else:
        confirmed=total_signups
        waitlisted=0
        status="Open"
    return confirmed, waitlisted, status
event_limit=int(input("Enter Event Capacity: "))
registrations_received=int(input("Enter Total Registrations Received: "))
confirmed_count, waitlist_count, current_status = registrations(event_limit, registrations_received)
print("Confirmed Registrations:", confirmed_count)
print("Waitlisted Users:", waitlist_count)
print("Registration Status:", current_status)
```
