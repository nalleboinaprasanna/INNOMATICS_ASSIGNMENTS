```python
#Storing credentials
user="admin"
password="1234"

#inputing data
entered_user=input("Username: ")
entered_pass=input("Password: ")

# 3. Validation Check
if entered_user==user and entered_pass==password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
#student marks in a list
marks=[45, 78, 90, 33, 60]

pass_count=0
fail_count=0

# Loop for marks
for score in marks:
    if score>=50:
        pass_count+=1  # Increment
    else:
        fail_count+=1  # Increment

print(f"Total Students Passed:{pass_count}")
print(f"Total Students Failed:{fail_count}")
```

```python
#uncleaned data
names=[" Alice ", "bob", " CHARLIE "]
cleaned_names=[]

#for loop for the data
for name in names:
    clean_data=name.strip().lower()
    cleaned_names.append(clean_data)

#Cleaned data
print("Cleaned List:", cleaned_names)
```

```python
#messages in the list
messages=["Hi", "Welcome to the platform", "OK"]
#checking the lenght of the messages
for msg in messages:
    length = len(msg)
    print(f"Message:'{msg}': Length:{length}")
#alert message for length above 10
    if length > 10:
        print("ALERT: This message is too long!")
```

```python
#data
data=["INFO", "ERROR", "WARNING", "ERROR"]
count=0

#searching ERROR in data
for entry in data:
    if entry=="ERROR":
        #counting the errors in list
        count=count+1
print(f"Total Errors Found: {count}")
```
