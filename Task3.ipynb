```python
#1
#employees data
employees={
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
# Finding the highest score
highest_score=max(employees.values())
#Creating an empty list to store the winners
top=[]
#Looping through the dictionary
for name in employees:
    if employees[name]==highest_score:
        top.append(name)
#Print the output
print("Top Performers Eligible for Bonus:", top)
print("Highest Score:", highest_score)
```

```python
#2
#give the imput query
sentence = input("Enter your search query: ")
#sentence to lower cases
sentence=sentence.lower()
#splitting the sentence
words = sentence.split()
#creating a dict for word count
word_counts = {}
#loop
for word in words:
    if word in word_counts:
        word_counts[word]=word_counts[word]+1
    else:
        word_counts[word]=1
#dict for words
keywords={}
for word in word_counts:
    if word_counts[word]>1:
        keywords[word]=word_counts[word]
#result
print("Keywords searched more than once:")
print(keywords)
```

```python
#3
#Defining the list
readings=[3, 4, 7, 8, 10, 12, 5]
#Creating an empty list
valid_readings=[]
#Looping through the list
for hour in range(len(readings)):
    reading=readings[hour]
    if reading % 2==0:
        # Creating a pair and adding it to our list
        pair=(hour, reading)
        valid_readings.append(pair)
#result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
#4
#mails
emails=[
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
#number of mails
total_emails=len(emails)
# 3. Creating an empty dictionary
counts={}
#looping
for email in emails:
    domain=email.split("@")[1]
    if domain in counts:
        counts[domain]=counts[domain]+1
    else:
        counts[domain]=1
#calucalting the percentage
for domain in counts:
    count=counts[domain]
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage}%")
```

```python
#5
#Defining the list
sales=[1200, 1500, 900, 2200, 1400, 3000]
total_sales=sum(sales)
number_of_days=len(sales)
average_sales=total_sales/number_of_days
limit=average_sales * 1.30
for i in range(len(sales)):
    current_sale=sales[i]
    if current_sale>limit:
        day_number=i+1
        print("Day", day_number, ":", current_sale)
```

```python
#6
#Defining the list
ids=["user1", "user2", "user1", "user3", "user1", "user3"]

#Creating an empty dict
id_counts={}
#looping
for uid in ids:
    if uid in id_counts:
        id_counts[uid]=id_counts[uid]+1
    else:
        id_counts[uid]=1
for uid in id_counts:
    # If the count is greater than 1, it means it is a duplicate
    if id_counts[uid] > 1:
        count_value = id_counts[uid]
        print(f"{uid} → {count_value} times")
```
