```python
#1
#function for recharge
def recharge(amount):
    #defining the plans
    valid_plans=[199, 299, 399, 599]
    if amount>=50 and amount in valid_plans:
        return True
    else:
        return False
#main
while True:
    print("Available Plans: 199, 299, 399, 599")
    #input
    user_amount=int(input("Enter recharge amount: "))
    #function call
    if recharge(user_amount):
        print(f"Success! Your recharge of â‚¹{user_amount} is processing.")
        break  # Exit the loop
    else:
        print("Invalid amount or plan. Please try again.")
        # Loop continues
```

```python
#2
def stock(inventory):
    #Loop through the dictionary
    for product, quantity in inventory.items():
        # Check if quantity is below the threshold of 15
        if quantity<15:
            print(f"Product: {product} | Stock: {quantity} -> Reorder Alert")
        else:
            print(f"Product: {product} | Stock: {quantity} -> Stock OK")

# main body for checking
inventory_data = {}
count=int(input("How many products to add? "))

for i in range(count):
    item_name=input(f"Enter name for product {i+1}: ")
    item_qty=int(input(f"Enter quantity for {item_name}: "))
    inventory_data[item_name]=item_qty

# Call the function with our dictionary
print("\n--- Inventory Report ---")
stock(inventory_data)
```

```python
#3
# Function to evaluate marks
def results(marks_list):
    total_marks=0
    for mark in marks_list:
        total_marks += mark
    count=len(marks_list)
    average=total_marks/count
    if average >= 50:
        return f"Average: {average:.2f} | Status: Pass"
    else:
        return f"Average: {average:.2f} | Status: Fail"

#main body
num_subjects = int(input("Enter number of subjects: "))
marks = []
for i in range(num_subjects):
    m = float(input(f"Enter marks for subject {i+1}: "))
    marks.append(m)

# Calling the function and printing the final result
if marks:
    result=results(marks)
    print(result)
else:
    print("No marks provided.")
```

```python
#4
# Function to calculate the total
def calculate(distance,peak):
    base_fare=50
    rate_per_km=12
    total=base_fare + (distance*rate_per_km)
    if peak=="yes":
        # 25% extra means multiplying by 1.25
        total=total*1.25
    return total

# Main loop
while True:
    print("CAB FARE\n")

    try:
        dist=float(input("Enter distance in km: "))
        peak=input("Is it peak hour? (yes/no): ").lower()

        # Calculate fare using our function
        final_bill=calculate(dist, peak)
        print(f"Your total fare is: â‚¹{final_bill:.2f}")
        repeat=input("Calculate another fare? (yes/no): ").lower()
        if repeat != "yes":
            print("Thank you for using our service!")
            break # Exit the loop

    except ValueError:
        print("Invalid input! Please enter a number for distance.")
```

```python
#5
# Function to calculate attendance
def eligibility(attendance):
    total_days=len(attendance)
    if total_days==0:
        return "No data provided"
    present_day=0
    # Loop through the list
    for status in attendance:
        if status=="P":
            present_day+=1
    # Calculate percentage
    percentage=(present_day / total_days) * 100
    if percentage >= 75:
        return f"Attendance: {percentage:.2f}% | Status: Eligible"
    else:
        return f"Attendance: {percentage:.2f}% | Status: Not Eligible"
user_entry=input("Enter attendance (P P A): ")
inP=user_entry.upper()
attendance_data=inP.split()
if attendance_data:
    result=eligibility(attendance_data)
    print(result)
else:
    print("No data entered!")
```

```python
# Function to check
def check_password(password):
    #Check minimum length
    if len(password) < 8:
        return "Weak: Password must be at least 8 characters long."

    has_digit=False
    has_special=False
    special_chars="@#$"
    for char in password:
        #Check for at least one digit
        if char.isdigit():
            has_digit=True
        if char in special_chars:
            has_special=True
    if has_digit and has_special:
        return "Strong Password: All security rules âœŒðŸ»"
    elif not has_digit:
        return "Weak: Add at least one digit (0-9)."
    else:
        return "Weak: Add at least one special character (@, #, or $)."

#main body
user_pwd=input("Enter a new password: ")

# Calling the function and printing the result
result=check_password(user_pwd)
print(result)
```
