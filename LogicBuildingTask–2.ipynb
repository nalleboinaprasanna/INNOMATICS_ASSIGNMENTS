```python
#1
# Input sentence
sentence=input(" ")
# Split the sentence into individual words
words=sentence.split()
# Convert the list of words into a set to remove duplicates
unique_words=set(words)
# Print the results
print(f"Unique words count: {len(unique_words)}")
print(f"Unique words: {unique_words}")
```

```python
#2
employees = {}
count = int(input("number of employees"))
for i in range(count):
    name=input(f"Enter name for employee {i+1}: ")
    salary=int(input(f"Enter salary for {name}: "))
    employees[name] = salary
if employees:
    highest_name = max(employees, key=employees.get)
    print(f"\nHighest Salary: {highest_name}-{employees[highest_name]}")
else:
    print("No employees.")
```

```python
#3
count=int(input("Numbers in the list: "))
numbers=[]
for i in range(count):
    val=int(input(f"Enter number: "))
    numbers.append(val)
if numbers:
    max=numbers[0]
    min=numbers[0]

    for num in numbers:
        if num>max:
            max=num
        if num<min:
            min=num

    print("List:", numbers)
    print("Max:", max)
    print("Min:", min)
else:
    print("The list is empty")
```

```python
#4
count=int(input("number of products: "))
prices=[]
for i in range(count):
    price=int(input(f"Enter price: "))
    prices.append(price)
above=0
for p in prices:
    if p>1000:
        above+=1
print("Products:", above)
```
